```python
import pandas as pd
data=pd.read_csv("E:\python programs\eye_cancer_patients - Copy (1).csv")
print(data)
```

```python
print(data.isnull().sum()) #null values checking
```

```python
data.columns #finding columns in dataset
```

```python
#loc and iloc are used to find value of particular row and column
#loc [row_inx, column_name ] and iloc [row_inx, column_inx]
print(data.loc[4990,"Cancer_Type"]) 
print(data.iloc[4990,3]) 
data.at[4990,"Cancer_Type"] #loc and at are same but at is faster than loc
```

```python
data.head(10) #to find first 10 rows of dataset
data.tail(10) #to find last 10 rows of dataset
data.shape() #to find number of rows and columns in dataset
```

```python
#GroupBy
data.groupby("Cancer_Type")["Age"].mean().astype(int)  #to find mean age of patients with different cancer types
```
